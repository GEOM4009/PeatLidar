```python
df = pd.read_csv("aggregated_stats.xls")
df
```

```python
# cmb means combined canopy height data for GEDI and ICESat
df["cmb_mean"] = df["h_canopy_mean"].combine_first(df["canopy_h_mean"])
df["cmb_min"] = df["h_canopy_min"].combine_first(df["canopy_h_mean"])
df["cmb_max"] = df["h_canopy_max"].combine_first(df["canopy_h_max"])
df["cmb_std"] = df["h_canopy_std"].combine_first(df["canopy_h_std"])
df
```

```python
# Calculate difference for canopy height mean, max, min, std. Concatenate into one dataframe
mean_diff = df.groupby("polygon_id")["cmb_mean"].apply(lambda x: x.max() - x.min())
max_diff = df.groupby("polygon_id")["cmb_max"].apply(lambda x: x.max() - x.min())
min_diff = df.groupby("polygon_id")["cmb_min"].apply(lambda x: x.max() - x.min())
std_diff = df.groupby("polygon_id")["cmb_std"].apply(lambda x: x.max() - x.min())
final_df = pd.concat([mean_diff, max_diff, min_diff, std_diff], axis=1)
final_df
```

```python
# Export as CSV
final_df.to_csv("comparison.csv", index=False)
```
