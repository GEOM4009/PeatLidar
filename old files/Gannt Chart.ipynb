```python
# Gannt Chart

df = pd.DataFrame([
    dict(Task="Report 1", Start='2026-01-14', Finish='2026-02-01', Resource="All", Completion_pct=100),
    dict(Task="Report 2", Start='2026-02-01', Finish='2026-02-22', Resource="All", Completion_pct=0),
    dict(Task="Report 3", Start='2026-02-22', Finish='2026-03-08', Resource="All", Completion_pct=0),
    dict(Task="Report 4", Start='2026-03-08', Finish='2026-03-29', Resource="All", Completion_pct=0),
    dict(Task="Testing", Start='2026-03-08', Finish='2026-03-25', Resource="All", Completion_pct=0),
    dict(Task="Finalize Code & Docs", Start='2026-03-25', Finish='2026-03-29', Resource="All", Completion_pct=0),
    dict(Task="Presentation", Start='2026-03-08', Finish='2026-03-31', Resource="All", Completion_pct=0),
    dict(Task="Final Report", Start='2026-03-08', Finish='2026-04-08', Resource="All", Completion_pct=0),
    dict(Task="Retrieve ICESat-2 Data", Start='2026-02-01', Finish='2026-03-25', Resource="Vincent", Completion_pct=0),
    dict(Task="Retrieve GEDI Data", Start='2026-02-01', Finish='2026-03-25', Resource="Vincent", Completion_pct=0),
    dict(Task="Convert Hdf5 Data", Start='2026-02-10', Finish='2026-03-25', Resource="Joshua", Completion_pct=0),
    dict(Task="Aggregation", Start='2026-02-10', Finish='2026-03-25', Resource="Ethan", Completion_pct=0),
    dict(Task="IceSat-2 & GEDI Comparison", Start='2026-02-10', Finish='2026-03-25', Resource="Joshua", Completion_pct=0),
    dict(Task="Take User Inputs", Start='2026-02-10', Finish='2026-03-25', Resource="Ethan", Completion_pct=0),
    dict(Task="Prepare Outputs", Start='2026-02-10', Finish='2026-03-25', Resource="Vincent", Completion_pct=0)
    
])

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", color="Resource")
fig.update_yaxes(autorange="reversed")
fig.show()

```
